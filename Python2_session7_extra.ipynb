````python
class Expr:
    def __call__(self, **context):
        pass

    def d(self, wrt):
        pass


class Const(Expr):
    def __init__(self, value):
        self.value = value

    def __call__(self, **context):
        return self.value

    def d(self, wrt):
        return Const(0)

    def __str__(self):
        return str(self.value)


class Var(Expr):
    def __init__(self, name):
        self.name = name

    def __call__(self, **context):
        return context[self.name]

    def d(self, wrt):
        if isinstance(wrt, Var) and self.name == wrt.name:
            return Const(1)
        else:
            return Const(0)

    def __str__(self):
        return self.name


V = Var
C = Const
x = V("x")
print(x)          # x
print(C(42))      # 42
print(C(42)())    # 42
print(C(42).d(x)())  # 0
print(x(x=42))       # 42
print(x.d(x)())      # 1
print(x.d(V("y"))())  # 0
````
