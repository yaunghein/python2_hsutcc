Factorial
10! = 10 _ 9 _ ... _ 1
10! = 10 _ 9!

9! = 9 \* 8!

2! = 2 \* 1!
1! = 1

`factorial(10) = 10 * factorial(9)`


````python
# 5! = 5 x 4 x 3 x 2 x 1
def recursive_factorial(x):
    if x == 1:
        return 1
    return x * recursive_factorial(x-1)


recursive_factorial(10)
````

````python
# 5! = 5 x 4 x 3 x 2 x 1
def interative_factorial(x):
    result = 1

    for number in range(1, x + 1):
        result *= number

    return result


interative_factorial(10)
````

````python
# %%timeit
import time

recursive_factorial(500)

t1 = time.time()
result = recursive_factorial(6)
t2 = time.time()
print('Time usage for recursive_factorial 6: ', t2-t1)
result = recursive_factorial(501)
t3 = time.time()
print('Time usage for recursive_factorial 500: ', t3-t2)
````

````python
# %%timeit
import time

interative_factorial(500)

t1 = time.time()
result = interative_factorial(6)
t2 = time.time()
print('Time usage for interative_factorial 6: ', t2-t1)
result = interative_factorial(501)
t3 = time.time()
print('Time usage for interative_factorial 500: ', t3-t2)
````

1. Write a Python function to find the max of any numbers of arguments (don't use the build in function `max()`)


````python
def get_max_number(*numbers):
    if not numbers:
        return None

    max_number = numbers[0]
    for number in numbers[1:]:
        if number > max_number:
            max_number = number

    return max_number


get_max_number(1, 2, 3, 4)
````

4. Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.


````python
def count_case(text: str):
    upper_count = 0
    lower_count = 0

    for character in text:
        if character.isupper():
            upper_count += 1
        elif character.islower():
            lower_count += 1

    return {'upper': upper_count, 'lower': lower_count}


count_case("Hello World")
````

6. Write a Python function to check whether a string is pangram or not.

Note : Pangrams are words or sentences containing every letter of the alphabet at least once.
For example : "The quick brown fox jumps over the lazy dog"


````python
import string


def is_pangram(text: str):
    result = True
    lower_text = text.lower()
    lower_letters = list(string.ascii_lowercase)

    for letter in lower_letters:
        if letter not in lower_text:
            result = False
            break

    return result


is_pangram('The quick brown fox jumps over the lazy dog')
````

7. Write a Python function to calculate the fibonacci number of index i (starts with 1). The function accepts the number as an argument and return a single number.
   The fibonacci sequence starts with 1,1,2,3,5,8,13,21,.... Therefore, fibonacci of index 6 is number 8.


````python
def get_fibonacci_number(number):
    sequence = []
    for index in range(number):
        if len(sequence) < 2:
            sequence.append(1)
        else:
            sequence.append(sequence[index - 1] + sequence[index - 2])
    return sequence[-1]


get_fibonacci_number(6)
````

8. Create a function which finds the length of each word in the phrase (broken by spaces) and return the values in a list. The function will have an input of a string, and output a list of integers.


````python
def word_lengths(phrase):
    words = phrase.split()
    lengths = []
    for word in words:
        lengths.append(len(word))
    return lengths


print(word_lengths("The quick brown fox"))
````

9. Write a function that accepts a list of tuples containing the name of the student and their score, sort them by the score from high to low.


````python
def sort_students_by_score(students):
    return sorted(students, key=lambda x: x[1], reverse=True)


students = [
    ("Alice", 88),
    ("Bob", 95),
    ("Charlie", 72),
    ("Diana", 95)
]

print(sort_students_by_score(students))
````

````python
students = [
    ("Alice", 88),
    ("Bob", 95),
    ("Charlie", 72)
]


def get_top_student(students):
    scores = list(map(lambda student: student[1], students))
    top_index = 0
    for index in range(1, len(scores)):
        if scores[index] > scores[top_index]:
            top_index = index
    return students[top_index][0]

# def get_top_student(students):
#     top_index = 0
#     for index in range(1, len(students)):
#         if students[index][1] > students[top_index][1]:
#             top_index = index
#     return students[top_index][0]


get_top_student(students)
````
